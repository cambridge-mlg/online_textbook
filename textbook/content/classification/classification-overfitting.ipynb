# Overfitting in classification

We have previously encountered [linear classification models](classification-logistic-regression-model.ipynb) and their [non-linear generalisation](classification-non-linear.ipynb). Both the linear and non-linear variants of these models can overfit to the training data. You may have noticed signs of this on some of the examples previously presented. In this section we will look at a couple of clear examples of this phenomenum using test-train splits to diagnose the problem as we did for [regression](../regression/regression-overfitting.ipynb).

```python
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Image, HTML, set_matplotlib_formats

set_matplotlib_formats('pdf', 'svg')
css_style = open('../../_static/custom_style.css', 'r').read()
HTML(f'<style>{css_style}</style>')

%config InlineBackend.figure_format = 'svg'
```

## Linear logistic classification

Although over-fitting is most problematic for non-linear models, it can still occur in linear models. The 2D Iris dataset we have used before is shown below with the data split into 75% for training and 25% for testing.

```python
# Load the Iris dataset
x = np.load('iris_inputs_full.npy')
y = np.load('iris_labels.npy')

# Keep only the first two classes -- discard the third class
x = x[np.where(np.logical_not(y == 2))[0]] 
y = y[np.where(np.logical_not(y == 2))[0]]

# Keep only the first two input features
x = x[:, 0:2] 

# Slice out training and test data
num_test = (x.shape[0]) // 4

x_train = x[num_test:]
y_train = y[num_test:]
x_test = x[:num_test]
y_test = y[:num_test]

# Indices for selecting training and test data
class_0_train = np.where(y_train == 0)
class_1_train = np.where(y_train == 1)
class_0_test = np.where(y_test == 0)
class_1_test = np.where(y_test == 1)

# Plot training and test data
plt.scatter(x_train[class_0_train, 0],
            x_train[class_0_train, 1],
            color='blue',
            marker='x',
            s=15,
            label='class 0, train')

plt.scatter(x_train[class_1_train, 0],
            x_train[class_1_train, 1],
            color='red',
            marker='x',
            s=15,
            label='class 1, train')

plt.scatter(x_test[class_0_test, 0],
            x_test[class_0_test, 1],
            color='blue',
            marker='o',
            s=20,
            label='class 0, test')

plt.scatter(x_test[class_1_test, 0],
            x_test[class_1_test, 1],
            color='red',
            marker='o',
            s=20,
            label='class 1, test')
plt.legend()

# Format plot
plt.title("Two dimensional binary Iris Dataset", fontsize=18)
plt.xlabel("$x_1$", fontsize=16)
plt.ylabel("$x_2$", fontsize=16)
plt.legend()
plt.show()
```

We will now run the code that fits the model using maximum-likelihood estimation through gradient ascent. Before running the code, take a moment to consider the following questions:

1. **Where will the decision boundary lie after running many iterations?** 
2. **What will the magnitude of the weights be?**
3. **What will be the held out log-likelihood of the resulting model**

$$\begin{align}
\mathcal{L} = \sum_{n=1}^{N_{\text{test}}} \log p(y_n^{\ast} | \mathbf{w},x_n^{\ast})
\end{align}$$

If the answer to this questions is not obvious, you may like to experiment with the code to allow, varying the number of iterations to form a picture about what's happening. 

```python
def sigmoid(a):
    return 1 / (1 + np.exp(-a))

def gradient_ascent(x_train, y_train, x_test, y_test, w0, num_steps, stepsize):
    
    # Add 1's to the inputs as usual, absorbing the bias term
    x_train = np.append(np.ones(shape=(x_train.shape[0], 1)), x_train, axis=1)
    x_test = np.append(np.ones(shape=(x_test.shape[0], 1)), x_test, axis=1)
    
    # Copy weights (to prevent changing init_weights as a side-effect
    w = w0.copy() 
    
    # Arrays for storing weights and log-liklihoods at each step
    w_history = []
    log_liks_train = []
    log_liks_test = []
    
    for n in range(num_steps):
        
        # Record current log-lik and current weights
        log_lik_train = np.sum(y_train * np.log(sigmoid(x_train.dot(w))) + \
                               (1 - y_train) * np.log(1 - sigmoid(x_train.dot(w))))
        
        log_lik_test = np.sum(y_test * np.log(sigmoid(x_test.dot(w))) + \
                              (1 - y_test) * np.log(1 - sigmoid(x_test.dot(w))))
        
        log_liks_train.append(log_lik_train)
        log_liks_test.append(log_lik_test)
        w_history.append(w.copy())
    
        # Compute (x^T x)
        sigma = sigmoid(x_train.dot(w))
        
        # Calculate gradient of log-likelihood w.r.t. w
        dL_dw = np.mean((y_train - sigma) * x_train.T, axis = 1) 
        
        # Update weights and repeat
        w += stepsize * dL_dw
    
    return np.array(w_history), np.array(log_liks_train), np.array(log_liks_test)
```

```python
# Use an initialisation close to the optimum, to speed up the optimisation.
# There is a unique optimum which gradient descent will find eventually,
# so poorer initialisations will only have us wait longer. NB there are three
# weights because there are two features plus a bias term.
w_init = [-4, 7, -11]

# Number of training steps and step size
num_steps = 2000
learn_rate = 5e-1

# Train logistic regression model
w_history, log_liks_train, log_liks_test = gradient_ascent(x_train,
                                                           y_train,
                                                           x_test,
                                                           y_test,
                                                           w_init,
                                                           num_steps,
                                                           learn_rate)
```

```python
# Inputs to plot at
x1 = np.linspace(4, 7.5, 100)
x2 = np.linspace(1.5, 5, 100)

# Grid of points to plot at
plot_grid = np.stack(np.meshgrid(x1, x2), axis=2)
plot_grid = plot_grid.reshape((-1, 2))
plot_grid = np.append(np.ones(shape=(plot_grid.shape[0], 1)),
                      plot_grid,
                      axis=1)

predictions = sigmoid(plot_grid.dot(w_history[-1]))
predictions = predictions.reshape(100, 100)

# Figure to plot on
fig = plt.figure(figsize=(10, 4), dpi=80)

# Subplot for the data and predictions
plt.subplot(1, 2, 1)

plt.contourf(x1, 
             x2,
             predictions, 
             cmap='coolwarm', 
             alpha=0.5, 
             origin='lower')

plt.scatter(x_train[class_0_train, 0],
            x_train[class_0_train, 1],
            color='blue',
            marker='x',
            s=10)
plt.scatter(x_train[class_1_train, 0],
            x_train[class_1_train, 1],
            color='red',
            marker='x',
            s=10)

plt.scatter(x_test[class_0_test, 0],
            x_test[class_0_test, 1],
            color='blue',
            marker='o',
            s=15)
plt.scatter(x_test[class_1_test, 0],
            x_test[class_1_test, 1],
            color='red',
            marker='o',
            s=15)

plt.title("Predictions", fontsize=18)
plt.xlabel("$x_1$", fontsize=16)
plt.ylabel("$x_2$", fontsize=16)

# Plot a circle around the misclassified point
plt.scatter([4.50747563],
            [2.30301912],
            color='green', 
            marker='o', 
            s=180,
            facecolors='none')

# Subplot for the train and test log-likelihoods
ax2 = plt.subplot(1, 2, 2)

# Plot train and test log-likelihoods
plt.plot(log_liks_train, color='black', label='Train log-likelihood')
plt.plot(log_liks_test, color='red', label='Test log-likelihood')

# Format plot
plt.title("Overfitting on Iris Dataset", fontsize=18)
plt.xlabel("$x_1$", fontsize=16)
plt.ylabel("$x_2$", fontsize=16)
plt.xticks(np.linspace(0, 2000, 5))
plt.yticks(np.linspace(-4., 0., 5))
plt.legend()

plt.tight_layout()
plt.show()
```

The training data above (crosses) are linearly separable. For this reason there are a set of maximum-likelihood solutions. Each solution places the linear decision boundary so that it divides the two classes (there are many ways to do this) and then makes the logistic function as steep as possible - i.e. a step function - so that the probability of each datapoint goes to 1 and the training log-likelihood goes to zero. Numerical problems may be encountered before this limit is reached as this behaviour causes the magnitude of the weights to go to infinity. 

Whilst the training likelihood of the maximum-likelihood solution is one, the test likelihood is zero. The blue test point in the bottom left (indicated by a green circle) is classified as being in class 1 which is incorrect. Worse still, it makes this prediction with probability 1 as the point is below the hard decision boundary, which zeros the likelihood. 

Clearly this is an instance of overfitting as the training likelihood is as high as it can possibly be and the test log-likelihood is as low as it can possibly be. The model is overconfident. Notice that this behaviour is very likely to happen for small numbers of datapoints which will often be linearly separable by chance. 

## Non-linear logistic classification

Let's consider a second example. Here we will use the 1D dataset below. 

```python
# Load one-dimensional version of the iris dataset
x = np.load('class_1d_inputs.npy')
y = np.load('class_1d_labels.npy')

x_train, x_test, y_train, y_test = x[:20], x[20:], y[:20], y[20:]

# Indices for selecting training and test data
class_0_train = np.where(y_train == 0)
class_1_train = np.where(y_train == 1)
class_0_test = np.where(y_test == 0)
class_1_test = np.where(y_test == 1)

# Figure on which to plot
fig = plt.figure(figsize=(10, 3), dpi=80)

plt.scatter(x_train[class_0_train,0],
            np.ones((1,np.size(class_0_train))),
            color='blue',
            marker='x',
            s = 15,
            label='class 0, train')

plt.scatter(x_train[class_1_train, 0], 
            np.ones((1,np.size(class_1_train))), 
            color='red', 
            marker='x', 
            s = 15,
            label='class 1, train')

# Plot test data
plt.scatter(x_test[class_0_test, 0],
            np.ones((1,np.size(class_0_test))),
            color='blue',
            marker='o',
            s=20,
            label='class 0, test')

plt.scatter(x_test[class_1_test, 0],
            np.ones((1,np.size(class_1_test))),
            color='red',
            marker='o',
            s=20,
            label='class 1, test')

# Format plot
plt.title("One dimensional binary Iris Dataset", fontsize=18)
plt.xlabel("$x_1$", fontsize=16)

# Remove y ticks, show legend
plt.gca().get_yaxis().set_ticks([])
plt.legend()

plt.show()
```

We will now fit a non-linear logistic classification model. We will use 100 Gaussian basis functions

$$\begin{align}
\phi_{d}(x) = \exp\left(-\frac{1}{2 \ell^2} ( \mathbf{x} - \mu_{d})^2\right)
\end{align}$$

evaluating the test and training likelihoods for  basis function length-scales that range from $\ell = 0.1$ to $\ell = 3$. **What do you expect the plot of the two likelihood functions to look like?**
**What will happen to the model at the long length-scales?**
**What will happen at the model at the short length-scales?**

```python
# Lengthscales to try out
num_lengthscales = 20
min_log_scale = -1.
max_log_scale = 0.5

lengthscales = np.logspace(min_log_scale,
                           max_log_scale,
                           num_lengthscales)

# Basis function locations
num_basis = 100
mu = np.linspace(-2.5, 4.0, num_basis)

# Number of optimisation steps and learning rate
num_steps = 5000
learn_rate = 1e-1

N_train = y_train.size
N_test = y_test.size

# Arrays to store training and test log likelihoods
ll_train = np.zeros((num_lengthscales, 1))
ll_test = np.zeros((num_lengthscales, 1))

for i in range(num_lengthscales):
        
    # Lengthscale to use
    lengthscale = lengthscales[i]
    phi_train = np.zeros((N_train, num_basis))
    phi_test = np.zeros((N_test, num_basis))

    # Compute values of basis functions at train and test data
    for n in range(N_train):
        for d in range(num_basis):   
            phi_train[n, d] = np.exp(-1 / (2 * lengthscale ** 2) * (mu[d] - x_train[n]) ** 2)

    for n in range(N_test):
        for d in range(num_basis):   
            phi_test[n, d] = np.exp(-1 / (2 * lengthscale ** 2) * (mu[d] - x_test[n]) ** 2)

    # Initialise weights
    w_init = np.zeros(shape=(num_basis+1,))
    w_history, log_liks_train, log_liks_test = gradient_ascent(phi_train,
                                                               y_train,
                                                               phi_test,
                                                               y_test,
                                                               w_init,
                                                               num_steps,
                                                               learn_rate)

    # Store train and testing log likelihoods
    ll_train[i] = log_liks_train[-1]
    ll_test[i] = log_liks_test[-1]
```

```python
# Plot results
plt.figure(figsize=(8, 4))

# Plot train and test log likelihoods
plt.plot(lengthscales,ll_train, color='black',label='Train $\mathcal{L}$')
plt.plot(lengthscales,ll_test, color='red',label='Test $\mathcal{L}$')

# Format plot
plt.title("Training and test log-likelihoods", fontsize=18)
plt.xlabel("Lengthscale $\\ell$", fontsize=16)
plt.ylabel("Log-likelihood $\mathcal{L}$", fontsize=16)
plt.gca().legend()
plt.tight_layout()
plt.show()
```

Notice the following:

1. **For small length-scales the training log-likelihood is at its highest**. As the length-scale is increased the training log-likelihood monotonically decreases. 

2. The test log-likelihood is always below the training log-likelihood. This is typical for any inferential procedure (even for Bayesian inference) as the model has been directly fit to the training data, so the predictions on these data are almost bound to be best. 

3. **For small length-scales the test log-likelihood is at its lowest**. As the length-scale is increased the test log-likelihood increases to a maximum around $\ell = 1.5$ and thereafter decreases slightly.

Let's try and understand this behaviour. Let's start by plotting the fit at the maximum of the **test likelihood** of the length-scale  $\ell \approx 1.5$:

```python
# Lengthscale and number of basis functions to use
lengthscale = 1.5
num_basis = 100

# Locations of basis functions
mu = np.linspace(-2.5, 4.0, num_basis)

# Numbers of training and test points
N_train = y_train.size
N_test = y_test.size

# Empty arrays for string the values of the basis functins
phi_train = np.zeros((N_train, num_basis))
phi_test = np.zeros((N_test, num_basis))

# Number of optimisation steps and learning rate
num_steps = 5000
learn_rate = 1e-1

for n in range(N_train):
    for d in range(num_basis):   
        phi_train[n,d] = np.exp(-1 / (2 * lengthscale ** 2) * (mu[d] - x_train[n]) ** 2)

for n in range(N_test):
    for d in range(num_basis):   
        phi_test[n,d] = np.exp(-1 / (2 * lengthscale ** 2) * (mu[d] - x_test[n]) ** 2)

# Initialise weights
w_init = np.zeros(shape = (num_basis + 1,))
w_history, log_liks_train, log_liks_test = gradient_ascent(phi_train,
                                                           y_train,
                                                           phi_test,
                                                           y_test,
                                                           w_init,
                                                           num_steps,
                                                           learn_rate)
```

```python
# Figure to plot on
plt.figure(figsize = (8, 4))

# Locations at which to evaluate the sigmoid
N_eval = 100
xs = np.linspace(mu[0], mu[-1], N_eval)

# Array for holding the values of the basis functions at the evaluation points
phi_eval = np.zeros((N_eval, num_basis))

for n in range(N_eval):
    for d in range(num_basis):   
        phi_eval[n, d] = np.exp(-1 / (2 * lengthscale ** 2) * (mu[d] - xs[n]) ** 2)

# Append ones to basis functions
phi_eval = np.append(np.ones(shape=(phi_eval.shape[0], 1)), phi_eval, axis=1)

# Plot training and test data
plt.scatter(x_train[class_0_train,0], 
            np.zeros((1, np.size(class_0_train))), 
            color='blue', 
            marker='x', 
            s=15,
            label='class 0, train')

plt.scatter(x_train[class_1_train, 0], 
            np.ones((1,np.size(class_1_train))), 
            color='red', 
            marker='x', 
            s=15,
            label='class 1, train')

plt.scatter(x_test[class_0_test, 0], 
            -0.1*np.ones((1,np.size(class_0_test))), 
            color = 'blue', 
            marker='o', 
            s=20,
            label='class 0, test')

plt.scatter(x_test[class_1_test, 0], 
            1.1*np.ones((1,np.size(class_1_test))), 
            color='red', 
            marker='o', 
            s=20,
            label='class 1, test')

# Plot model predictions
plt.plot(xs, sigmoid(phi_eval.dot(w_history[-1,:])), color = 'black', zorder = 3, label = 'p(y|x,w)')

# Format plot
plt.title("Fit for the maximum-likelihood lengthscale", fontsize=18)
plt.xlabel("$x$", fontsize=16)
plt.ylabel("$p(y = 1|\mathbf{w}, x)$", fontsize=16)

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.gca().legend()
plt.tight_layout()
plt.show()
```

Here we have separated the test and the train data to help assess the model. This looks like a reasonable fit that takes into account the indistinct bounday between the two classes. Plots for larger length-scales look similar to this, but they over-smooth the boundary slightly, reducing the associated test likelihood. Now let's find out what's happening at those very low length-scales which actually **maximise the training likelihood** by plotting the fit for $\ell = 0.1$.

```python
# Lengthscale and number of basis functions to use
lengthscale = 0.1
num_basis = 100

# Locations of basis functions
mu = np.linspace(-2.5, 4.0, num_basis)

# Numbers of training and test points
N_train = y_train.size
N_test = y_test.size

# Empty arrays for string the values of the basis functins
phi_train = np.zeros((N_train, num_basis))
phi_test = np.zeros((N_test, num_basis))

# Number of optimisation steps and learning rate
num_steps = 5000
learn_rate = 1e-1

for n in range(N_train):
    for d in range(num_basis):   
        phi_train[n,d] = np.exp(-1 / (2 * lengthscale ** 2) * (mu[d] - x_train[n]) ** 2)

for n in range(N_test):
    for d in range(num_basis):   
        phi_test[n,d] = np.exp(-1 / (2 * lengthscale ** 2) * (mu[d] - x_test[n]) ** 2)

# Initialise weights
w_init = np.zeros(shape = (num_basis + 1,))
w_history, log_liks_train, log_liks_test = gradient_ascent(phi_train,
                                                           y_train,
                                                           phi_test,
                                                           y_test,
                                                           w_init,
                                                           num_steps,
                                                           learn_rate)
```

```python
# Figure to plot on
plt.figure(figsize=(8, 4))

N_eval = 500
xs = np.linspace(mu[0], mu[-1], N_eval)

phi_eval = np.zeros((N_eval,num_basis))
for n in range(N_eval):
    for d in range(num_basis):   
        phi_eval[n,d] = np.exp(-1/(2*lengthscale**2)*(mu[d] - xs[n])**2)

# Append ones to basis functions
phi_eval = np.append(np.ones(shape = (phi_eval.shape[0], 1)), phi_eval, axis=1)

# Plot training and test data
plt.scatter(x_train[class_0_train,0], 
            np.zeros((1,np.size(class_0_train))), 
            color='blue', 
            marker='x', 
            s=15,
            label='class 0, train')

plt.scatter(x_train[class_1_train, 0], 
            np.ones((1,np.size(class_1_train))), 
            color='red', 
            marker='x', 
            s=15,
            label='class 1, train')

plt.scatter(x_test[class_0_test, 0],
            -0.1*np.ones((1,np.size(class_0_test))),
            color='blue',
            marker='o',
            s=20,
            label='class 0, test')

plt.scatter(x_test[class_1_test, 0],
            1.1*np.ones((1,np.size(class_1_test))),
            color='red',
            marker='o',
            s=20,
            label='class 1, test')

plt.plot(xs,
         sigmoid(phi_eval.dot(w_history[-1,:])),
         color='black',
         zorder=3,
         label='p(y|x,w)')

# Format plot
plt.title("Fit of short length-scale", fontsize=18)
plt.xlabel("$x$", fontsize=16)
plt.ylabel("$y$", fontsize=16)

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.gca().legend()
plt.tight_layout()
plt.show()
```

We can now see the source of the bad behaviour: the model is contorting itself so that it assigns a probability that is close to one for each of the training points. The model can now do this as the length-scale of the basis functions is very shot. The predictions on the test data are poor. This is second example of over-fitting and an example of **a failure of maximum likelihood training**. 

## Summary

Overfitting occurs in classification when the model can contort itself to assign probability one to each training datapoint. This then results in confident predictions that are often wrong when evaluated on the test set. This behaviour can manifest itself in different ways depending on the data, model, and the method of fitting the parameters:

1. Linear classification methods trained by maximising the training likelihood will overfit in cases where the training data classes are perfectly separable by a linear decision boundary. 
2. Non-linear logistic classification with RBF basis functions trained by maximising the training likelihood will overfit when a large number of basis functions are used and the length-scales are short. The resulting model will be overl complex.

Overfitting can be identified using train / test splits of the data. When the training likelihood is increasing and the test likelihood is decreasing it is likely that overfitting is occurring.

In [section 3.5](classification-regularisation.ipynb) we will look at using MAP inference, rather than maximum likelihood estimation, can regularise the solution and mitigate overfitting.
